```python
import torch
import torchvision
from torchvision import datasets, transforms
import matplotlib.pyplot as plt
import torch.optim as optim
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
from tqdm import tqdm
```

```python
train = datasets.MNIST("",train=True,download=False,
                      transform= transforms.Compose([transforms.ToTensor()]))
test = datasets.MNIST("",train=False,download=False,
                      transform= transforms.Compose([transforms.ToTensor()]))
```

```python
trainset = torch.utils.data.DataLoader(train, batch_size=10, shuffle=True)
testset = torch.utils.data.DataLoader(test, batch_size=10, shuffle=True)
```

```python
for data in testset:
    for i in range(len(data)):
        
        x,y=data[0][i],data[1][i]
        plt.imshow(x.reshape(28,28),cmap='gray')
        plt.show()
        print(x.shape)
        #print(y)
    
    break
```

```python
class Net(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(1,16,5,1)
        self.conv2 = nn.Conv2d(16,64,5,1)
        #self.conv3 = nn.Conv2D(64,128,3)
        self.fc1 = nn.Linear(4*4*64,500)
        self.fc2 = nn.Linear(500,10)
        
    def forward(self,x):
        # TODO: add dropout
        x = F.max_pool2d(F.relu(self.conv1(x)),2,2)
        x = F.max_pool2d(F.relu(self.conv2(x)),2,2)
        x=x.view(-1,4*4*64)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return F.softmax(x,dim=1)
```

```python
class RMSELoss(nn.Module):
    def __init__(self, eps=1e-6):
        super().__init__()
        self.mse = nn.MSELoss()
        self.eps = eps
        
    def forward(self,yhat,y):
        loss = torch.sqrt(self.mse(yhat,y) + self.eps)
        return loss
```

```python
net=Net()
optimizer=optim.Adam(net.parameters(),lr=0.001)
epochs=1
for epoch in range(epochs):
    for data in tqdm(trainset):
        train_x , train_y = data[0],torch.Tensor(np.eye(10)[data[1]])
        net.zero_grad()
        output=net(train_x)
        #print(output.shape)
        #print(train_y.shape)
        rmse=RMSELoss()
        loss=rmse(output,train_y)
        loss.backward()
        optimizer.step()
print(loss)
```

```python
correct=0
total=0

with torch.no_grad():
    for data in tqdm(testset):
        test_x , test_y = data[0],torch.Tensor(np.eye(10)[data[1]])
        output=net(test_x)
        for idx, i in enumerate(output):
            if torch.argmax(i) == torch.argmax(test_y[idx]):
                correct += 1
            total += 1
print('Accuracy: {}'.format(round(correct/total,3)))
```
